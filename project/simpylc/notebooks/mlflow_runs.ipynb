```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# load datasets
df = pd.read_csv('runs.csv')
df2 = pd.read_csv('runs2.csv')
df.describe()
df2.describe()
```

```python
# only keep the columns that are interesting
df = df.loc[:,['batch_size', 'batchnorm', 'epochs',
               'layers', 'lr', 'num_layers', 'relu', 'sensor', 'mse_list']]
df2 = df2.loc[:,['batch_size', 'batchnorm', 'epochs',
               'layers', 'lr', 'num_layers', 'relu', 'sensor', 'mse_list']]
```

```python
# sort by cost (ascending) and print the first 20 rows
df.sort_values('mse_list').head(50)
```

```python
# sort by cost (ascending) and print the first 20 rows
df2.sort_values('mse_list').head(50)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("MSE (sensor=lidar, relu=True)")
plt.subplots_adjust(hspace=0.7)
sns.boxplot(ax=ax1, x=df['layers'], y=df["mse_list"], palette=["m"])
ax1.set_xticklabels(df['layers'].unique(), rotation=90)
ax1.set_ylabel('MSE')
ax1.set_xlabel('Layers')

sns.boxplot(ax=ax2, x=df['batch_size'], y=df["mse_list"], palette=["m"])
ax2.set_ylabel('MSE')
ax2.set_xlabel('Batch size')

sns.boxplot(ax=ax3, x=df['lr'], y=df["mse_list"], palette=["m"])
ax3.set_ylabel('MSE')
ax3.set_xlabel('Learning rate')

sns.boxplot(ax=ax4, x=df['epochs'], y=df["mse_list"], palette=["m"])
ax3.set_ylabel('MSE')
ax3.set_xlabel('Epochs')

plt.show()
```

```python
# group by layers and calculate average loss
df_layers = df2.groupby(['layers'])['mse_list'].mean()
df_bs = df2.groupby(['batch_size'])['mse_list'].mean()
df_relu = df2.groupby(['relu'])['mse_list'].mean()
df_bn = df2.groupby(['batchnorm'])['mse_list'].mean()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
plt.subplots_adjust(hspace=0.7)
ax1.bar(df_layers.index, df_layers)
ax1.set_xticklabels(df_layers.index, rotation=90)
ax1.set_ylabel('Mean MSE')
ax1.set_xlabel('Layers')

ax2.bar(df_bs.index, df_bs)
ax2.set_ylabel('Mean MSE')
ax2.set_xlabel('Batch size')

ax3.bar(df_relu.index, df_relu)
ax3.set_ylabel('Mean MSE')
ax3.set_xticks([0.0, 1.0], ["False", "True"])
ax3.set_xlabel('Relu')

ax4.bar(df_bn.index, df_bn)
ax4.set_ylabel('Mean MSE')
ax4.set_xticks([0.0, 1.0], ["False", "True"])
ax4.set_xlabel('Batchnorm')

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("MSE (sensor=lidar, lr=0.00001, epochs=500)")
plt.subplots_adjust(hspace=0.7)
sns.boxplot(ax=ax1, x=df2['layers'], y=df2["mse_list"], hue=df2["relu"], palette=["m", "g"])
ax1.set_xticklabels(df_layers.index, rotation=90)
ax1.set_ylabel('MSE')
ax1.set_xlabel('Layers')

sns.boxplot(ax=ax2, x=df2['batch_size'], y=df2["mse_list"], hue=df2["relu"], palette=["m", "g"])
ax2.set_ylabel('MSE')
ax2.set_xlabel('Batch size')

sns.boxplot(ax=ax3, x=df2['relu'], y=df2["mse_list"], palette=["m"])
ax3.set_ylabel('MSE')
ax3.set_xticks([0.0, 1.0], ["False", "True"])
ax3.set_xlabel('Relu')

sns.boxplot(ax=ax4, x=df2['batchnorm'], y=df2["mse_list"], hue=df2["relu"], palette=["m", "g"])
ax4.set_ylabel('MSE')
ax4.set_xticks([0.0, 1.0], ["False", "True"])
ax4.set_xlabel('Batchnorm')

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("MSE (sensor=lidar, lr=0.00001, epochs=500, relu=True)")
plt.subplots_adjust(hspace=0.7)
sns.boxplot(ax=ax1, x=df2[df2['relu'] == True]['layers'], y=df2[df2['relu'] == True]["mse_list"], palette=["m"])
ax1.set_xticklabels(df2['layers'].unique(), rotation=90)
ax1.set_ylabel('MSE')
ax1.set_xlabel('Layers')

sns.boxplot(ax=ax2, x=df2[df2['relu'] == True]['batch_size'], y=df2[df2['relu'] == True]["mse_list"], palette=["m"])
ax2.set_ylabel('MSE')
ax2.set_xlabel('Batch size')

sns.boxplot(ax=ax3, x=df2[df2['relu'] == True]['batchnorm'], y=df2[df2['relu'] == True]["mse_list"], palette=["m"])
ax3.set_ylabel('MSE')
ax3.set_xticks([0.0, 1.0], ["False", "True"])
ax3.set_xlabel('Batchnorm')

fig.delaxes(ax4)

plt.show()
```
